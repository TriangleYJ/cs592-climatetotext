```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "4"

import json
from bert_score import score
from tqdm import tqdm

# Load data from jsonl file
predictions = []
references = []

with open(data, 'r', encoding='utf-8') as f:
    for line in tqdm(f, desc="Loading data"):
        item = json.loads(line.strip())
        predictions.append(item['infer'])
        references.append(item['answer'])

print(f"Loaded {len(predictions)} samples")
print(f"\nExample prediction: {predictions[0][:200]}...")
print(f"\nExample reference: {references[0][:200]}...")
```

```python
# Calculate BERTScore
P, R, F1 = score(predictions, references, lang='en', verbose=True, model_type='microsoft/deberta-xlarge-mnli', batch_size=4)
print(f"\n{'='*50}")
print(f"BERTScore Results:")
print(f"{'='*50}")
print(f"Precision: {P.mean():.4f} (±{P.std():.4f})")
print(f"Recall:    {R.mean():.4f} (±{R.std():.4f})")
print(f"F1:        {F1.mean():.4f} (±{F1.std():.4f})")
print(f"{'='*50}")
```

```python
# Detailed statistics
import numpy as np

print("\nDetailed Statistics:")
print(f"\nPrecision - Min: {P.min():.4f}, Max: {P.max():.4f}, Median: {P.median():.4f}")
print(f"Recall    - Min: {R.min():.4f}, Max: {R.max():.4f}, Median: {R.median():.4f}")
print(f"F1        - Min: {F1.min():.4f}, Max: {F1.max():.4f}, Median: {F1.median():.4f}")

# Show top 5 and bottom 5 samples by F1 score
indices = F1.argsort()
print("\n" + "="*50)
print("Top 5 Best F1 Scores:")
print("="*50)
for i in indices[-5:].flip(0):
    print(f"\nF1: {F1[i]:.4f}")
    print(f"Prediction: {predictions[i][:150]}...")
    print(f"Reference:  {references[i][:150]}...")

print("\n" + "="*50)
print("Bottom 5 Worst F1 Scores:")
print("="*50)
for i in indices[:5]:
    print(f"\nF1: {F1[i]:.4f}")
    print(f"Prediction: {predictions[i][:150]}...")
    print(f"Reference:  {references[i][:150]}...")
```
